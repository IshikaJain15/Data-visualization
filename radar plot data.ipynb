```python
sales_4['SalesOrderCreationDate'] = pd.to_datetime(sales_4['SalesOrderCreationDate'])

# Filter sales data for a particular month and year
desired_year = 2022
desired_month = 1

filtered_sales = sales_4[(sales_4['SalesOrderCreationDate'].dt.year == desired_year) &
                          (sales_4['SalesOrderCreationDate'].dt.month == desired_month)]

# Calculate summary statistics
total_orders = filtered_sales['SalesOrder'].nunique()
total_order_quantity = filtered_sales['OrderQuantity'].sum()
average_order_quantity = filtered_sales['OrderQuantity'].mean()

# Display summary
print(f"Summary for {desired_month}/{desired_year}:")
print(f"Total orders: {total_orders}")
print(f"Total order quantity: {total_order_quantity}")
print(f"Average order quantity: {average_order_quantity}")
```

```python
unique_materials = sales_4['MaterialKey'].unique()

# Initialize an empty dictionary to store material-wise quantity ordered for each month
material_quantity_by_month = {}

# Iterate through each month and year combination
for year in range(sales_4['SalesOrderCreationDate'].dt.year.min(), sales_4['SalesOrderCreationDate'].dt.year.max() + 1):
    for month in range(1, 13):
        # Filter sales data for the current month and year
        filtered_sales = sales_4[(sales_4['SalesOrderCreationDate'].dt.year == year) &
                                  (sales_4['SalesOrderCreationDate'].dt.month == month)]
        
        # Initialize an empty dictionary to store material-wise quantity ordered for the current month
        month_material_quantity = {}
        
        # Iterate through each unique material
        for material in unique_materials:
            # Filter sales data for the current material
            material_sales = filtered_sales[filtered_sales['MaterialKey'] == material]
            # Calculate total quantity ordered for the current material in the current month
            total_quantity = material_sales['OrderQuantity'].sum()
            # Store the total quantity ordered for the current material in the dictionary
            month_material_quantity[material] = total_quantity
        
        # Store the material-wise quantity ordered for the current month in the main dictionary
        material_quantity_by_month[(year, month)] = month_material_quantity

# Display the results
for year_month, material_quantity in material_quantity_by_month.items():
    year, month = year_month
    print(f"Year: {year}, Month: {month}")
    for material, quantity in material_quantity.items():
        print(f"Material: {material}, Quantity Ordered: {quantity}")
    print()
```

```python
import pandas as pd

# Assuming metrics_data is a DataFrame containing the metrics data
# Load metrics_data into a DataFrame (replace this with your actual data loading code)
metrics_data = pd.read_excel("metrics_plant4_material2.xlsx")

# Define a function to normalize the data between -1 and 1
def normalize(data):
    min_val = data.min()
    max_val = data.max()
    return -1 + 2 * ((data - min_val) / (max_val - min_val))

# Exclude Year and Month columns from normalization
normalized_data = metrics_data.copy()
normalized_data[['GoodsReceiptDelay', 'ArrivalDelay', 'ShipmentDelay', 'DeliveryDelay', 
                 'Sales Percentage Monthly Increase', 'Inventory Proportion in Sales']] = \
    normalized_data[['GoodsReceiptDelay', 'ArrivalDelay', 'ShipmentDelay', 'DeliveryDelay', 
                     'Sales Percentage Monthly Increase', 'Inventory Proportion in Sales']].apply(normalize, axis=0)

# Write the normalized DataFrame to an Excel file
normalized_data.to_excel("normalized_metrics_plant4_material2.xlsx", index=False)

print("Excel file 'normalized_metrics_data.xlsx' has been created with the normalized metrics data.")
```
